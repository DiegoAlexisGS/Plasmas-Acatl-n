```python
import numpy as np
from numpy import pi
import pandas as pd
import matplotlib.pyplot as plt
```

Consideramos las ecuaciones 
$$
\omega=\frac{arctan(\frac{v\delta_t}{\delta}tan(k\delta))}{\delta_t}
$$
y
$$
\gamma=-\frac{ln[cos^2(k\delta)+(\frac{v\delta_t}{\delta})^2sin^2(k\delta)]}{2t}
$$

```python
def omega(d,dt,v,k):
    return((np.arctan((v*dt/d)*np.tan(k*d)))/dt)
def gamma(d,dt,v,k):
    return((-np.log(np.cos(k*d)**2+(v*dt/d)**2*np.sin(k*d)**2))/2*dt)
```

```python
y=[]
y1=[]
y2=[]
z=[]
z1=[]
z2=[]
q=1
x=[0,pi/8*q,2*pi/8*q,4*pi/8*q]
for i in x:
    y.append(omega(1,1,2,i))
    y1.append(omega(1,1,0,i))
    y2.append(omega(2,2,1,i))
    z.append(gamma(1,1,0.5,i))
    z1.append(gamma(1,1,0,i))
    z2.append(gamma(1,1,0.1,i))
```

```python
fig, ax=plt.subplots()
ax.plot(x,y,color="blue")
ax.plot(x,y1,color="green")
ax.plot(x,y2,color="red")
ax.set_xlabel("k")
ax.set_ylabel("$\omega$")
ax.legend()
plt.show()
```

```python
fig, ax=plt.subplots()
ax.plot(x,z,color="blue")
ax.plot(x,z1,color="green")
ax.plot(x,z2,color="red")
ax.set_xlabel("k")
ax.set_ylabel("$\gamma$")
ax.legend()
plt.show()
```
