# Método de Euler

```python
from numpy import *
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
```

Consideramos la ecuacíon $$\frac{du(x)}{dx}=\frac{-u}{\tau}$$
con la condición inicial $u(0)=1$.

La solución analitíca de la ecuación diferencial es $$u(x)=u_0e^{\frac{-x}{\tau}}$$
donde por la condición inicial $u_0=1$

```python
"""
Algoritmo de Euler
"""

a=0 #Extremo inferior del intervalo
b=2 #Extremo superior del intervalo
N=10 #Entero N 
A=0.5 #condición inicial y(a)=A

def euler(a,b,N,A,f):
    #Paso 1
    listax=[]
    listay=[]
    h=(b-a)/N
    t=a
    w=A
    #Paso 2
    for i in range(1,N+1):
        #Paso 3
        w=w+h*float(f(t,w))
        t=a+i*h
        listax.append(t)
        listay.append(w)
    return(listax,listay)
        
```

```python
#Función analítica

m=linspace(0,2,20,endpoint=True)
result=[]
for i in m:
    result.append(e**(-i/1))
    

"""
N=20
T=1 #tau
"""
def ser(t,u):
    T=1
    return(-u/T)
x,t=euler(0,2,20,1,ser)

"""
N=20
T=5 #tau
u_0=2
"""
def ser(t,u):
    T=5
    return(-u/T)

x1,t1=euler(0,2,20,1,ser)
"""
N=20
T=10 #tau
"""
def ser(t,u):
    T=10
    return(-u/T)

x2,t2=euler(0,2,20,1,ser)
```

```python
fig, ax=plt.subplots()
ax.plot(x,t,color="blue",label="T=1,u_0=1")
ax.plot(x1,t1,color="red",label="T=5, u_0=1")
ax.plot(x2,t2,color="green",label="T=10, u_0=1")
ax.plot(m,result,color="purple",label="e^(-x)")
ax.set_xlabel("t")
ax.set_ylabel("f(t)")
ax.legend()
plt.show()
```

```python
fig, axes=plt.subplots(1,3,figsize=(12,4))
ax=axes[0].plot(x,t,color="blue",label="T=1,u_0=1")
ax=axes[1].plot(x1,t1,color="red",label="T=5, u_0=1")
ax=axes[2].plot(x2,t2,color="green",label="T=10, u_0=1")
ax=axes[0].set_xlabel("t")
ax=axes[0].set_ylabel("f(t)")
ax=axes[1].set_xlabel("t")
ax=axes[1].set_ylabel("f(t)")
ax=axes[2].set_xlabel("t")
ax=axes[2].set_ylabel("f(t)")
ax=axes[0].legend()
ax=axes[1].legend()
ax=axes[2].legend()
plt.show()
```

# Caso con la ecuación de movimiento armonico simple

Consideramos \begin{equation}\frac{d^2x}{dt^2}=-\omega_0^2 x\end{equation}
con las condiciones iniciales $x(0)=1$ $\frac{dx(0)}{dt}=0$
Pasando la ecuación original a un sistema de ecuaciones tenemos 
$$
u=x
$$
$$
v=x'
$$
Entonces 
$$
u'=x'=v
$$
$$
v'=x''=-\omega_0 x
$$
Por lo tanto el sistema obtenido es
$$
\frac{dx(t)}{dt}=v
$$
$$
\frac{dv(t)}{dt}=-\omega_0^2x
$$

```python
def euler2(a,b,N,A,A1):
    h=(b-a)/N
    t_0=a
    x_0=A
    v_0=A1
    lista1=[]
    lista2=[]
    lista1.append(t_0)
    lista2.append(x_0)
    for i in range(1,N+1):
        t_0=a+i*h
        x_0=x_0+h*v_0
        v_0=v_0+h*(-x_0)
        lista1.append(t_0)
        lista2.append(x_0)
    return(lista1,lista2)
    
```

Resolvemos para la ecuación 
$$
\frac{d^2x}{dt^2}=-x
$$
con las condiciones iniciales $x(0)=1$, $x'(0)=2$ en el intervalo $[0,2]$

```python

t0,x0=euler2(0,2,20,1,2)
df=pd.DataFrame(x0,t0)
df.head()
```

```python
def analitica(t):
    return(2*sin(t)+cos(t))
puntos=linspace(0,2,20,endpoint=True)
lt=[]
for i in puntos:
    lt.append(analitica(i))
```

```python
fig, ax=plt.subplots()
ax.plot(t0,x0,color="blue",label="númerica")
ax.plot(puntos,lt,color="red",label="analítica")
ax.set_xlabel("t")
ax.set_ylabel("f(t)")
ax.legend()
plt.show()
```
