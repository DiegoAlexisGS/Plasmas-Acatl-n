# Método de Euler

```python
from numpy import *
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
"""
Algoritmo de Euler
"""

a=0 #Extremo inferior del intervalo
b=2 #Extremo superior del intervalo
N=10 #Entero N 
A=0.5 #condición inicial y(a)=A

def euler(a,b,N,A,f):
    #Paso 1
    listax=[]
    listay=[]
    h=(b-a)/N
    t=a
    w=A
    #Paso 2
    for i in range(1,N+1):
        #Paso 3
        w=w+h*float(f(t,w))
        t=a+i*h
        listax.append(t)
        listay.append(w)
    return(listax,listay)
        
```

```python
"""
N=20
T=1 #tau
"""
def ser(t,u):
    T=1
    return(-u/T)

a=0 #Extremo inferior del intervalo
b=2 #Extremo superior del intervalo
N=20 #Entero N 
A=1 #condición inicial y(a)=A

x,t=euler(a,b,N,A,ser)
```

```python
"""
N=20
T=10 #tau
"""
def ser(t,u):
    T=10
    return(-u/T)

a=0 #Extremo inferior del intervalo
b=2 #Extremo superior del intervalo
N=20 #Entero N 
A=1 #condición inicial y(a)=A

x2,t2=euler(a,b,N,A,ser)
```

```python
fig, ax=plt.subplots()
ax.plot(x,t,color="blue",label="T=1,u_0=1")
ax.plot(x1,t1,color="red",label="T=5, u_0=1")
ax.plot(x2,t2,color="green",label="T=10, u_0=1")
ax.set_xlabel("t")
ax.set_ylabel("f(t)")
ax.legend()
plt.show()
```
